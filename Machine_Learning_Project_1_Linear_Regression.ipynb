```python
#Importing some functions
from csv import reader
from random import randrange
```

```python
#Function to load csv:
def load_csv(filename):
	dataset = list() 
	with open(filename, 'r') as file:
		csv_reader = reader(file)
		for col in csv_reader:
			if not col:
				continue
			dataset.append(col) #Fetching the value in the dataset from file
	return dataset
```

```python
#String column to float operaton:
def str_column_to_float(dataset, column):
	for col in dataset:
		col[column] = float(col[column].strip())   #Using the Strip Function for this operation
```

```python
#Loading the required CSV file from the system:      
filename = 'data2.csv'
dataset = load_csv(filename)
len(dataset)              #Length of the dataset
```

```python
for i in range(len(dataset[0])):
	str_column_to_float(dataset, i)     #Converting string column to float values
```

```python
#Initializing Input and Ouput Features of Dataset:
X = [row[1:9] for row in dataset]
Y = [row[0] for row in dataset]
```

```python
#Scaling the data as there was a lot of difference in the scales of the parameters:
for i in range(len(X[0])):
    for index, item in enumerate(X):
        try:
            X[index][i] = int(item[i])
        except Exception as e:
            X[index][i] = 0


    max_val = max([row[i] for row in X])
    min_val = min([row[i] for row in X])
    for index, item in enumerate(X):
        try:
            X[index][i] = (item[i]-min_val)/(max_val-min_val) #Formula to get the scaling between 0 and 1
        except Exception as e:
            X[index][i] = 0
```

```python
# Split the data into a train and test set:
def train_test_split(XX, split = 0.8):  #split is 0.8 to keep the division as 80% training data and 20% for testing.
	X_train = list()
	X_train_size = split * len(XX)      #Splitting into training and test set
	dataset_copy = list(XX)           #Keeping a copy of the data in XX
	while len(X_train) < X_train_size:
		index = randrange(len(dataset_copy))
		X_train.append(dataset_copy.pop(index))   #Now the training data set is fetched into the copied dataset.
	return X_train, dataset_copy
```

```python
#Assigning Input and Output Features:
X1 = [row[0] for row in X_train] #No of Bedrooms
X2 = [row[1] for row in X_train] #Sqft_Living
X3 = [row[2] for row in X_train] #Sqft_Lot
X4 = [row[3] for row in X_train] #No of Floors
X5 = [row[4] for row in X_train] #Condition
X6 = [row[5] for row in X_train] #Grade
X7 = [row[6] for row in X_train] #Sqft_Living15
X8 = [row[7] for row in X_train] #Sqft_Lot15
```

```python
#Initializing with random guesses for the values of our parameters thetas:
t0 = 4;t1 = 5;t2 = 3;t3 = 9;t4 = 6;t5 = 2;t6 = 1;t7 = 8;t8 = 5; 
```

```python
#Performing Linear Regression for the Training Set:
def simple_linear_regression(X_train):        #Using X_train as it is training
	predictions = []
	for col in X_train:                      #Taking data from X_train
		y_pred = t0 + t1 * col[0]+t2 * col[1]+t3 * col[2]+t4 * col[3]+t5 * col[4] + t6*col[5] + t7*col[6] + t8*col[7]
		predictions.append(y_pred)
	return predictions
```

```python
#Hypothesis value for initial values of our parameters:
prediction = simple_linear_regression(X_train)
```

```python
m = len(X_train)         #Identifying m as length of X_train, to be used.  
```

```python
#Cost Function for calculating cost or mean squared error:
def cost_function(X,Y):
    cost=0
    summ=0
    for i in range(m):
        summ=(1/(2*m))*((X[i]-Y[i])**2)
        cost=cost+summ

    return cost
```

```python
#Displaying the result of cost function:
cost=cost_function(prediction,Y_train)
cost
```

```python
#Initializing the learning rate 'alpha':
alpha = 0.000000005
```

```python
def Gradient(predict,Y,p0,p1,p2,p3,p4,p5,p6,p7,p8,X1,X2,X3,X4,X5,X6,X7,X8):  #Gradient function for calculating Grad. Descent:
    predictions = []
    for i in range(100):          #No. of epochs or iterations
        cost0 = 0
        cost1 = 0
        cost2 = 0
        cost3 = 0
        cost4 = 0
        cost5 = 0
        cost6 = 0
        cost7 = 0
        cost8 = 0
        for i in range(len(predict)):                  #Loop for calculating the cost regarding each feature
            gradient=((predict[i]-Y[i]))
            cost0=cost0+gradient

            gradient1=gradient*X1[i]
            cost1=cost1+gradient1

            gradient2=gradient*X2[i]
            cost2=cost2+gradient2



            gradient3=gradient*X3[i]
            cost3=cost3+gradient3



            gradient4=gradient*X4[i]
            cost4=cost4+gradient4


            gradient5=gradient*X5[i]
            cost5=cost5+gradient5
            
            gradient6=gradient*X6[i]
            cost6=cost6+gradient6
            
            gradient7=gradient*X7[i]
            cost7=cost7+gradient7
            
            gradient8=gradient*X8[i]
            cost8=cost8+gradient8
        p0=p0-(alpha*cost0/m)                      #The values of our parameters, 'thetas'
        p1=p1-(alpha*cost1/m)
        p2=p2-(alpha*cost2/m)
        p3=p3-(alpha*cost3/m)
        p4=p4-(alpha*cost4/m)
        p5=p5-(alpha*cost5/m)
        p6=p6-(alpha*cost6/m)
        p7=p7-(alpha*cost7/m)
        p8=p8-(alpha*cost8/m)
        for col in X_train:
            y_pred = p0 + p1 * col[0]+p2 * col[1]+p3 * col[2]+p4 * col[3]+p5 * col[4] + p6*col[5] + p7*col[6] + p8*col[7]
            predictions.append(y_pred)
        c=cost_function(predictions,Y)                   #Calculating cost-function for training
        if c<15686236403461306630413031438745122:
            break
        else:
            Gradient(predictions,Y,p0,p1,p2,p3,p4,p5,p6,p7,p8,X1,X2,X3,X4,X5,X6,X7,X8)
    return p0,p1,p2,p3,p4,p5,p6,p7,p8,c #Returning values of thetas and cost function
    
```

```python
#The updated values of thetas and cost function:
Gradient(prediction,Y_train,t0,t1,t2,t3,t4,t5,t6,t7,t8,X1,X2,X3,X4,X5,X6,X7,X8)
```

```python
#Simple linear Regression for hypothesis for Test set:
def simple_linear_regression(X_test):
	predictions = []
	for col in X_test:                  #Using X_test for hypothesis and updated values of thetas
		y_pred = t0 + t1 * col[0]+t2 * col[1]+t3 * col[2]+t4 * col[3]+t5 * col[4] + t6*col[5] + t7*col[6] + t8*col[7]
		predictions.append(y_pred)
	return predictions
```

```python
#Hypothesis of Test-set:
prediction_test = simple_linear_regression(X_test)
```

```python
#Calculating the accuracy for the test set:
n=len(Y_test)
def accuracy(yt,p):
    sum=0
    for i in range(len(yt)):
        sum=sum+abs((yt[i]-p[i]/yt[i]))
    m=(sum/n)
    return m
```

```python
#Accuracy using test set:
acc=accuracy(Y_test,prediction_test)
```
